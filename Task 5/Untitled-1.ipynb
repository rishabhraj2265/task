```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('nyc-rolling-sales.csv')
```

```python
df.columns = df.columns.str.strip().str.upper()
```

```python
numeric_columns = ['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE']
for col in numeric_columns:
    df[col] = df[col].replace({',': ''}, regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df = df.dropna(subset=['SALE PRICE'])
```

```python
df.head
```

```python
df['BOROUGH_BIN'] = df['BOROUGH']
```

```python
for col in ['LAND SQUARE FEET', 'GROSS SQUARE FEET']:
    df[col] = df.groupby('BOROUGH_BIN')[col].transform(lambda x: x.fillna(x.median()))
```

```python
for col in ['LAND SQUARE FEET', 'GROSS SQUARE FEET']:
    df[col] = df[col].fillna(df[col].median())
```

```python
df = df.drop(columns=['BOROUGH_BIN'])
```

```python
for col in ['LAND SQUARE FEET', 'GROSS SQUARE FEET', 'SALE PRICE']:
    df[col] = df[col].clip(lower=0)
```

```python
print("After cleaning:")
print(df.isna().sum())
print(f"Remaining rows: {len(df)}")
```

```python
df.to_csv('nyc_property_sales_cleaned.csv', index=False)
```
